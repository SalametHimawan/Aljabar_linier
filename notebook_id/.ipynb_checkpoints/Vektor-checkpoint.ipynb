
## Apa itu Vektor?

Blok penyusun dasar untuk aljabar linier adalah vektor, secara umum ada tiga interpretasi vektor yang berbeda tetapi terkait. Interpretasi berdasarkan perspektif fisika, perspektif ilmu komputer, dan perspektif matematikawan.

1. Perspektif Fisika, bahwa vektor adalah panah yang menunjuk dalam ruang. Vektor memiliki panjang dan arah, tetapi selama keduanya sama kita dapat memindahkannya dan itu masih vektor yang sama.
2. Perspektif Ilmu Komputer, bahwa vektor adalah urutan daftar angka.
3. Abstraksi Matematikawan, matematika menggeneralisasi kedua pandangan ini:
$\mathbf{u}+\mathbf{v}$, $\alpha\mathbf{v}$.

## Visualisasi vektor
### Sistem koordinat

Koordinat suatu vektor adalah sepasang angka yang berisi informasi vektor itu dari titik asal, ke ujungnya. Angka pertama merupakan informasi seberapa jauh berjalan di sepanjang sumbu x, dengan angka positif menunjukkan gerakan ke kanan dan angka negatif menunjukkan gerakan ke kiri, dan angka kedua memberi informasi seberapa jauh untuk kemudian berjalan sejajar dengan sumbu y, dengan angka positif menunjukkan gerakan ke atas, dan angka negatif menunjukkan gerakan ke bawah.

### Operasi Vektor

**Penjumlahan Vektor**, dan **Perkalian vektor dengan skalar**

Visualisasi dari penjumlahan dua vektor 

$$
   \mathbf{a} = \left[ \begin{array}{c} -2 \\ 1  \end{array} \right], \quad  
   \mathbf{b} = \left[ \begin{array}{c} 1 \\ -3  \end{array} \right] 
$$

Kita dapat memvisualisasikan dengan cara: gambarkan vektor $\mathbf{a}$ dari origin; selanjutnya gambarkan vektor $\mathbf{b}$ . Selanjutnya gambarkan vektor $\mathbf{b}$, yaitu merupakan $\mathbf{a} + \mathbf{b}$.


```python
# vector addition
a = numpy.array((-2,1))
b = numpy.array((1,-3))
origin = numpy.array((0,0))

vectors = [a, b, a+b]
tails   = [origin, a, origin]
plot_vector(vectors, tails)
pyplot.title("Penjumlahan vektor dengan koordinat $(-2, 1)$ dan $(1,-3)$.\n");
```

Penjumlahan $\mathbf{a} + \mathbf{b}$. Dapat dituliskan sebagai berikut:

$$
   \left[ \begin{array}{c} -2 \\ 1  \end{array} \right] +  
   \left[ \begin{array}{c} 1 \\ -3  \end{array} \right] =
   \left[ \begin{array}{c} -2+1 \\ 1-3  \end{array} \right]
$$


Perkalian vektor dengan skalar. Apabila Kita mengalikan vektor dengan suatu nilai skalar, seperti mengalikannya dengan $2$, maka panjang vektor tersebut akan menjadi $2$ lebih panjang. 

Contohnya Kita mengalikan $2$ dengan vektor $\mathbf{c} = \left[ \begin{array}{c} 2 \\ 1  \end{array} \right]$, maka akan terlihat seperti berikut:

```python
# vector scaling
c = numpy.array((2,1))
vectors = [c, 2*c]
plot_vector(vectors)
pyplot.title("Perkalian vektor $(2,1)$ dengan nilai skalar $2$.");
```

Dapat dilihat sebagai berikut:

$$
  2\cdot\left[ \begin{array}{c} 2 \\ 1  \end{array} \right] =
  \left[ \begin{array}{c} 2\cdot 2 \\ 2\cdot 1  \end{array} \right]
$$
